```python
import matplotlib.pyplot as plt

# Ensure 'issue_d' is datetime and extract year
my_data['issue_d'] = pd.to_datetime(my_data['issue_d'], errors='coerce')
my_data['issue_year'] = my_data['issue_d'].dt.year

# Remove rows where year is missing
my_data = my_data[my_data['issue_year'].notna()]

# Calculate total loans per year
total_loans_by_year = my_data.groupby('issue_year').size()

# Define the statuses to track
statuses = ['Late (31-120 days)', 'Late (16-30 days)', 'Charged Off']
colors = ['orange', 'gold', 'darkred']
labels = ['Late (31-120 days)', 'Late (16-30 days)', 'Charged Off']

# Plot each status
plt.figure(figsize=(10, 6))
for status, color, label in zip(statuses, colors, labels):
    count_by_year = my_data[my_data['loan_status'] == status].groupby('issue_year').size()
    perc_by_year = (count_by_year / total_loans_by_year * 100).fillna(0).sort_index()
    plt.plot(perc_by_year.index.astype(int), perc_by_year.values, marker='o', color=color, label=label)

# Final touches
plt.title('Yearly Percentage of Late and Charged Off Loans')
plt.xlabel('Year')
plt.ylabel('Percentage of Loans (%)')
plt.legend(title='Loan Status')
plt.xticks(sorted(my_data['issue_year'].unique().astype(int)))
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
